```python
#7.3
import pandas as pd
"""Create a student with grades and id and stuff"""
class Student():
    def __init__(self,name,ID):
        self.ID = ID
        self.name = name
        self.grades = pd.Series(dtype=float)

    def addas(self, assign, grade):
        self.grades.loc[assign] = grade

    def allgrades(self):
        self.grades.plot.bar(title="Grades: " + self.name + ', ' + self.ID)


```

```python
#7.3a
stu1 = Student("Rishi", 'rs15')
stu1.addas("HW01", 91)
stu1.addas("HW02", 98)
stu1.addas("HW03", 89)
stu1.addas("HW04", 88)
stu1.allgrades()
```

```python
stu2 = Student("notRishi", 'nr16')
stu2.addas("HW01", 70)
stu2.addas("HW02", 100)
stu2.addas("HW03", 85)
stu2.addas("HW04", 55)
stu2.allgrades()
```

```python
stu3 = Student("Somebdyelse", 'se99')
stu3.addas("HW01", 99)
stu3.addas("HW02", 99)
stu3.addas("HW03", 99)
stu3.addas("HW04", 99)
stu3.allgrades()
```

```python
#7.6
class Circle():
    def __init__(self, rad):
        self.rad = rad

    def area(self):
        return 3.14 * (self.rad**2) 
    
    def perimeter(self):
        return 2 * self.rad * 3.14
```

```python
circ = Circle(3)
print(circ.area())
print(circ.perimeter())
```

```python
#7.7
class Rectangle():
    
    def __init__(self, length, width):
        self.length = length
        self.width  = width
    
    def area(self):
        return self.length * self.width
```

```python
rect = Rectangle(5,8)
rect.area()
```
